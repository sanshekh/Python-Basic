```python
Inventory= {'Men': ['Shirt','Suit','Trouser','Jeans'], 'Women': ['Saree','Dress_suit','Handbag','Jewellery'],'Home_appliances': ['TV','Refrigerator'],'Kids':['Toys','Bags','Books']}
```

```python
def Inventory_check(x):
    for category in Inventory:
        if x in Inventory[category]:
            return True
        else:
            return False
```

```python
condition=True
while condition:
    item_name = input('Enter Name of the item that you want to buy?')
    if(Inventory_check(item_name)):
        print('Item available')
    else:
        print('Sorry!Item Not Available.')
    option = input('Do you want to continue shopping?')
    condition = True if option.lower() == 'yes' else False
```
